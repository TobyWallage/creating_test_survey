```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageDraw
from astropy.table import Table
from IPython.display import display
```

```python
image = Image.open('IDL_0305060101.png') 
display(image)
```

```python
print(image)
```

```python
input_table = Table.read('Python_final_source_summaries.dat', format='ascii')
```

```python
t_image = image.transpose(Image.Transpose.FLIP_TOP_BOTTOM)
for row in input_table:
    x, y = row['x'], row['y']
    id = row['id']

    bbox = (x-40, y-40, x+40, y+40)
    region = t_image.crop(bbox)

    region_scaled = region.resize((400,400))

    draw = ImageDraw.Draw(region_scaled)

    width, height = region_scaled.size

    cent_x, cent_y = width//2, height//2

    draw.line((cent_x,0,cent_x,height), fill=(192,192,192))
    draw.line((0, cent_y,width, cent_y), fill=(192,192,192))

    # display(region_scaled)

    region_scaled.save(f"./XAPA_IMAGES/{id}_XAPA_COMP_ext_IDL_img.png")
```

```python
image.close()
```
